```python
import pandas as pd
import mindsdb
from mindsdb import Predictor
```

# In this work,I made a desicion for preprocessing data, just remove nan data if total is greater than thresold ( in here is 0.7)
Got : 0.22150 ( Rank: 4072/ 4730)

```python
train=pd.read_csv("./train.csv")
test=pd.read_csv("./test.csv")
```

```python
train.select_dtypes(include='object').head()
```

```python
train.head()
```

```python
total = train.isnull().sum().sort_values(ascending=False)
```

```python
total.head(20)
```

```python
missing_data = pd.concat([total], axis=1, keys=['Total'])
```

```python
train_new = train.drop((missing_data[missing_data['Total'] > 37]).index,1)
```

```python
train_new.head()
```

```python
test_new = test.drop((total[total> 37]).index,1)
```

```python
test_new.head()
```

```python
Predictor(name='home_rentals_price').learn(
    to_predict='SalePrice', # the column we want to learn to predict given all the data in the file
    from_data=train_new # the path to the file where we can learn from, (note: can be url)
)
```

```python
f = open('result1.csv', 'w')
for i in range(0,len(test) + 1):
    a = test_new.iloc[i].to_dict()
    _id = a['Id']
    del a['Id']
    result = Predictor(name='home_rentals_price').predict(when=a)
    f.write(str(_id))
    f.write(',')
    f.write(str(result[0]['SalePrice']))
    f.write('\n')
    print(str(_id),str(result[0]['SalePrice']))
```

```python
f.close()
```
